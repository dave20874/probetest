```python
%matplotlib inline

import csv
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


```

Read data from probes.csv file into x, y, mean.

```python

x = []
y = []
mean = []
stddev = []
with open("probes.csv") as csv_file:
    reader=csv.reader(csv_file, delimiter=",")
    line = 0
    for row in reader:
        if line == 0:
            # print(f'Column names are {", ".join(row)}')
            pass
        else:
            x.append(float(row[0]))
            y.append(float(row[1]))
            mean.append(float(row[2]))
            stddev.append(float(row[6]))
        line += 1
```

Plot the mean surface as a function of position.

```python
fig = plt.figure(figsize=(16,8))
ax = fig.gca(projection='3d')
ax.plot_trisurf(x, y, mean, cmap=cm.ocean, linewidth=0.2)
plt.show()
```

So the surface measured by the diff IR sensor varies by over 1.5mm in areas smaller than 5x5mm.  This means Z probing can't be done consistently.  Now the difficulties with inconsistent print starts and bed leveling make sense.

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.plot_trisurf(x, y, stddev, cmap=cm.ocean, linewidth=0.2)
plt.show()
```

The standard deviation looks random.  The values are pretty small, 0.005mm being the max.  So the height measurements above are very reliable.
